```python
import seaborn as sns
import pandas
import os

os.getcwd()
#df = pandas.read_csv("version_vector_formatted.csv")
df = pandas.read_csv("version_vector_formatted.csv")
#df = pandas.read_csv("csv_format/version_vector_formatted.csv")

df['algorithm'] = 'version_vector'
df['sites']=df['sites'].astype(int)
print(df)
plot_exec_time = sns.lineplot(
    data=df, x="sites", y="exec", hue="algorithm", err_style="band"
)
plot_exec_time.set(title='Execution time plot')
plot_exec_time.set_xlabel("Num of sites", fontsize = 12)
plot_exec_time.set_ylabel("Execution time", fontsize = 12)
plot_exec_time.axes.set_ylim(0,20000)

```

```python
plot_icd_count = sns.lineplot(
    data=df, x="sites", y="icd", hue="algorithm", err_style="band"
)
plot_exec_time.set(title='Num of inconsistency detected plot')
plot_exec_time.set_xlabel("Num of sites", fontsize = 12)
plot_exec_time.set_ylabel("Num of inconsistency detected", fontsize = 12)
plot_exec_time.axes.set_ylim(0,100)
```
